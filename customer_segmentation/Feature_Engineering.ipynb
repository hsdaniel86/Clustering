```python
#date= [x for x in dt.datetime("%d-%m-%Y")]
data= pd.read_csv("data/marketing_campaign.csv", sep= "\t", parse_dates= ["Dt_Customer"])
data
```

```python
data.dtypes
```

```python
data.drop(["Z_CostContact", "Z_Revenue"], axis= 1, inplace= True)
data.head()
```

```python
data.rename(columns={'MntWines': 'Wines', 'MntFruits': 'Fruits','MntMeatProducts': 'Meat', 
                     'MntFishProducts': 'Fish', 'MntSweetProducts': 'Sweet','MntGoldProds': 'Gold', 
                     'NumDealsPurchases': 'DealsPurchases', 'NumWebPurchases': 'WebPurchases',
                     'NumCatalogPurchases': 'CatalogPurchases', 'NumStorePurchases': 'StorePurchases',
                     'NumWebVisitsMonth': 'WebVisitsMonth','AcceptedCmp3': 'Offer_3', 'AcceptedCmp4': 'Offer_4'
                     , 'AcceptedCmp5': 'Offer_5', 'AcceptedCmp1':'Offer_1', 'AcceptedCmp2': 'Offer_2'}, 
            inplace=True)
```

```python
# Creating numeric type variable 
data["Education_N"]= data["Education"].map({"Graduation": 2, "PhD": 3, "Master": 4, 
                                            "2n Cycle": 5, "Basic": 1})
```

```python
# Creating numeric type variable
data["Marital_Status_N"]= data["Marital_Status"].map({"Married": 1, "Together": 2, "Single": 3, "Divorced": 4, 
                                                    "Widow": 5, "Alone": 6, "Absurd": 7})
```

```python
# genereting Two new variables from the Dt_Customer variable 
data["Year"]= data["Dt_Customer"].dt.year
data["Month"]= data["Dt_Customer"].dt.month
data.head()
```

```python
# Generating subset for clustering, segmentation by value spent on products
data_products= data.iloc[:, 9:15]
data_products.head()
```

```python
# Including ID column
data_products["ID"]= data["ID"]
cols= data_products.columns.tolist()
cols= cols[-1:] + cols[:-1]
cols
```

```python
# Saving the dataframe in format
data_products.to_csv("data/data_products.csv",index=False)
```

```python
# Generating subset, personal data variables
data_personal= data.iloc[:, 0:9]
data_personal
```

```python
data_personal[["Education_N", "Marital_Status_N", "Year", "Month"]]= data.iloc[:, 27:]
data_personal["Complain"]= data.Complain
data_personal
```

```python
data_personal.to_csv("data/data_personal.csv", index= False)
```

```python
# Creating subset with variables from places where customers buy the most
data_purchase= data.iloc[:, 15: 20]
data_purchase
```

```python
# Including ID column
data_purchase["ID"]= data.ID
data_purchase= data_purchase.reindex(columns=["ID", "DealsPurchases", "WebPurchases", "CatalogPurchases",
                                              "StorePurchases", "WebVisitsMonth"])
data_purchase
```

```python
data_purchase.to_csv("data/data_purchase.csv", index=False)
```

```python
# generating subset of Offers
data_offer= data.iloc[:, 20:25]
data_offer[["ID", "Response"]]= data[["ID", "Response"]]
data_offer
```

```python
data_offer= data_offer.reindex(columns=["ID", "Offer_1", "Offer_2", "Offer_3", 
                                        "Offer_4", "Offer_5", "Response"])
data_offer
```

```python
data_offer.to_csv("data/data_offer", index= False)
```
